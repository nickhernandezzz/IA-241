```python
import pandas
```

```python
df = pandas.read_excel('s3://hernandez-ia241/COVID-19-geographic-disbtribution-worldwide.xlsx')
df
```

# Top 10 Countries with the Highest Number of Cases

```python
sum_death_by_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_death_by_country.nlargest(10).plot.bar()
```

# Belgium Data

```python
Belgium_data = df.loc[df['countriesAndTerritories'] == 'Belgium']
Belgium_data[:10] 
```

# Number of Cases Related to Deaths in Belgium

```python
Belgium_data.plot.scatter(x='cases',y='deaths')
```

# Average Number of Deaths per Day

```python
avg_deaths_per_day = df.groupby('dateRep').mean()['deaths']

avg_deaths_per_day.plot()
```
